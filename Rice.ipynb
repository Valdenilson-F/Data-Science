## **Read Me**
This dataset comprises 981 accessions characterized by 10 morphological traits, curated and conserved by the AfricaRice Genebank as of September 2021. The data were originally downloaded from Genesys-PGR, an open-source platform for plant genetic resources.

**Species Included**
Oryza sativa L. and Oryza glaberrima Steud.

**raits Evaluated**
Ten morphological/agronomic traits were evaluated (see documentation or dataset header for details).

### **Traits observed**
**Plant Type** 

Culm Angle - rice direct-seeded in rows.

**Leaf blade length**

Measure the highest leaf below the flag leaf (penultimate leaf) on the main culm from the ligule to the tip of the blade on five representative plants. Calculate average to nearest centimeter.

**Leaf blade width**

Measure at the widest portion of the penultimate leaf (i.e. highest leaf below the flag leaf) on the main culm on five representative plants. Calculate average to nearest cm. Oryza sativa leaf can be coded as narrow (< 1 cm), internediate (1-2 cm) or broad (>2 cm).

**Panicle length**

Length of main axis of panicle measured from the panicle base to the tip. Record the average of five representative plants. Stage: 7 days after anthesis or upon full panicle exsertion.

**Ligule length**

Ligule length in millimeter

**Ligule Shape**

**Ligule shape**

**Culm Diameter**

Culm diameter of basal internode at reproductive in millimeter

**Culm length**

Measured from ground level to the base of the panicle. Record the average of five actual measurements, to the nearest cm.

**Total number of tillers**

Count tillers with panicles and those without panicles on 5 samples

**Number of fertile tillers**

Count only the tillers with panicles on 5 samples.

## **The analisys:**

import the libraries:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr
```

uploading data:

```python
data=pd.read_csv('Data/Data.csv')
```

confering traits of columns:

```python
data
```

checking our data it is possible to see that some columns have a comma to separate the decimals and this will make them be read as a string

```python
data.describe()
```

then we will select the numeric columns of the dataframe and then apply the following code:

```python
cols= ['Leaf blade length [cm]', 'Leaf blade width [cm]',
       'Culm length [cm]', 'Panicle length [cm]', 'Ligule length [mm]',
       'Culm diameter [mm]','Total number of tillers ', 'Number of fertile tillers ']

data[cols]=data[cols].replace(',','.', regex=True)
data[cols]=data[cols].astype(float)
```

when describing our data again you can see that it selects all the numeric columns now

```python
data.describe()
```

At this point we need to transform the columns of our interest into a dataframe to create our graphs without major problems.

```python
df=data[cols]
df
```

It is always very interesting at the beginning of a multivariate analysis to apply point data to check the distribution of the data, see outliers and predictability.

```python
sns.pairplot(data[cols], diag_kind='kde')
plt.figure(figsize=(12,6))
plt.show()
```

A quick look at this matrix of graphs shows how the relationship between "Total number of tillers" and "Number of fertile tillers" distributes its values ââaround an imaginary curve, which is a clear sign that the data is correlated.

If in doubt, you can apply Pearson's correlation

```python
corr, pval = pearsonr(df['Total number of tillers '], df['Number of fertile tillers '])
print(f"Pearson correlation: {corr:.2f} (p-value: {pval:.6f})")
```

We can see that they have a positive correlation (if one increases the other increases too) of 77% and a p-value of 0 (meaning that there is no probability of this being due to chance)

```python
df = data[['Total number of tillers ', 'Number of fertile tillers ']].dropna()

sns.jointplot(
    data=df,
    x='Total number of tillers ', 
    y='Number of fertile tillers ',
    kind='reg',
    scatter_kws={'alpha':0.3, 'color':'#6666ff'},
    line_kws={'color':'#FF4D00', 'linewidth':1},
    
    
)

plt.xlabel('Total number of tillers ')
plt.xticks(np.arange(0,90,10))
plt.ylabel('Number of fertile tillers ')
plt.yticks(np.arange(0,110,10))
plt.grid(True, zorder=1)
plt.suptitle('Relationship between Vegetative Capacity and Productive Potential', fontsize=14)
plt.show()
```

When viewing the information in this graph, it becomes intuitive to see that rice reaches its maximum production capacity between 10-20 seeds, as this is where maximum vegetative use is concentrated.

based on this, we can ask ourselves 'what other characteristics are correlated' and we can create a 'heatmap' that answers us graphically

```python
df = df.apply(pd.to_numeric, errors='coerce')
pearson_corr=df.corr(method='pearson')

sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', center=0)
plt.title('CorrelaÃ§Ã£o de Pearson para Caracteristicas do Oryza Sativa')
plt.show()
```

Among the other characteristics that have a strong positive correlation, we can highlight 'culm length (cm)' with 'Leaf blade length (cm), 'leaf blade width (cm)' with 'leaf blade length (cm)' and 'culm length (cm) with leaf blade width (cm)'.
